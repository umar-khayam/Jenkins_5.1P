**(f) Create a new qualitative variable, called Elite, by binning the Top10perc variable into two groups based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50%.**

**college['Elite'] = pd.cut(college['Top10perc'], [0,0.5,1], labels=['No', 'Yes'])**

**Use the value_counts() method of college['Elite'] to see how many elite universities there are. Finally, use the boxplot() method again to produce side-by-side boxplots of Outstate versus Elite.**

```python
# divide tha top10perc into two parts and make new variable name Elite 
# Assign No where top10perc is less 50 else Yes

college['Elite'] = pd.cut(college['Top10perc'], [0,50,100], labels=['No', 'Yes'])
```

**2. What is the relationship of student enrollment with the total expenses of study, does more students want to study in the institute where expenses are low?.**

```python
# add the expenses personal and books and make a new variable Total expense
college['Total Expense'] = college['Personal'] + college['Books']
```

```python
import seaborn as sns
sns.scatterplot(data=college, x="Enroll", y="Total Expense")
plt.title('Relationship Between Student Enrollment Numbers and Total Expenses')
plt.xlabel('Enrollment Numbers')
plt.ylabel('Total Expenses ($)')
plt.show()
```

```python
boxplot = college.boxplot(column='Total Expense', by='Elite')
boxplot.set_title('Expenses for elite and non-elite institute')
boxplot.set_xlabel('Elite')
boxplot.set_ylabel('Total Expenses ($)')
plt.suptitle('')
plt.show()
```

Yes, mostly students want to be in the cheap institute, but the institute which are elite are more expensive than non-elite,
